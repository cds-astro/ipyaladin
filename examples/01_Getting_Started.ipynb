# Getting Started

Displays an interactive sky view of the [DECam Plane Survey](http://decaps.skymaps.info/).

```python
from ipyaladin import Aladin
```

```python
aladin = Aladin(
    survey="http://alasky.cds.unistra.fr/DECaPS/DR1/color/",
    show_coo_grid=True,
    target="galactic center",
    coo_frame="galactic",
    fov=40,
    height=600,
)
aladin
```
